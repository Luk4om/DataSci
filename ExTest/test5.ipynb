```python
dic = {'day':['Sun','Mon','Tue','Wed','Thu','Fri','Sat'] ,
       'temp':[24,25,29,35,23,20,23]}
df = pd.DataFrame(dic)
x = df['day']
y = df['temp']
plt.plot(x, y)
```

```python
dic = {'day':['Sun','Mon','Tue','Wed','Thu','Fri','Sat'] ,
       'temp1':[24,25,29,35,23,20,23] ,
       'temp2':[10,14,9,12,11,15,17]}
df = pd.DataFrame(dic)
x = df['day']
y = df['temp1']
z = df['temp2']
```

```python
df
```

```python
plt.plot(df['day'],df['temp1'], 's:r', label = 'temp1')
plt.plot(df['day'],df['temp2'], 'o--b', label = 'temp2')
plt.legend(loc = 'best')
# plt.legend(loc = 'upper left')
```

```python
fig, axes = plt.subplots( ncols = 2, nrows = 2 )
axes[0,0].plot(x, y, 's:r')
axes[0,1].plot(x, z, 'o--b')
axes[1,0].plot(y, x, 'p-k')
axes[1,1].plot(z, x, '*-.g')
axes[0,0].set_title("week1")
axes[0,1].set_title("week2")
axes[1,0].set_title("week3")
axes[1,1].set_title("week4")
fig.tight_layout()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x, y, 'r--')
axes1.set_title('Axes 1')
axes1.set_xlabel('X Axis')
axes1.set_ylabel('Y Axis')

axes2 = fig.add_axes([1.2,0,1,1])
axes2.plot(x, z, 'b--')
axes2.set_title('Axes 2')

axes3 = fig.add_axes([0,1.2,1,1])
axes3.plot(z, x, 'y--')
axes3.set_title('Axes 3')

axes4 = fig.add_axes([1.2,1.2,1,1])
axes4.plot(y, x, 'g--')
axes4.set_title('Axes 4')
```
