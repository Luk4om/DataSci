```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(1,21)
y = np.random.randint(1,100,20)
```

```python
plt.scatter(x,y,marker = 'o',c = 'r',s = 50,ec = 'k',lw = 2)
```

```python
plt.figure(figsize = [6,4])
x = ['Mon','Tue','Wed','Thu','Fri']
y = [50,70,35,90,40]
plt.bar(x,y,color = 'b',alpha = 0.5,hatch = '\/-|',lw = 2,ec = 'k',ls = ':')
# plt.plot(x,y)
plt.title('Week')
```

```python
plt.figure(figsize = [6,4])
x = ['Mon','Tue','Wed','Thu','Fri']
y = [50,70,35,90,40]
plt.barh(x,y,color = 'r',alpha = 0.5,hatch = '\/-|',lw = 2,ec = 'k',ls = ':')
# plt.plot(x,y)
plt.title('Week')
```

```python
x = np.arange(1,6)
y1 = np.random.randint(1,50,5)
y2 = np.random.randint(1,50,5)
y3 = np.random.randint(1,50,5)
```

```python
plt.figure(figsize = [6,4])
bw = 0.2
plt.bar(x,y1,width = 0.2,label = 'A')
plt.bar(x+0.2,y2,width = 0.2,label = 'B')
plt.bar(x+0.2*2,y3,width = 0.2,label = 'C')
plt.xticks(x+bw,['2000','2001','2002','2003','2004'])
plt.legend(loc = 'best')
plt.show()
```

```python
data = np.random.randint(1,100,200)
```

```python
data = [50,500,100,80]
x = ['Chrome','Firefox','Opera','Safari']
```

```python
plt.pie(data,labels = x,explode = [0.3,0,0.2,0.1],shadow = True,autopct = '%.1f%%')
plt.show()
```

```python
dic = {"Math" : [10,40,78,55,60],
       "Eng" : [99,23,56,87,65],
       "Thai": [78,50,60,34,25]
       }
df = pd.DataFrame(dic)
data = [df['Math'],df['Eng'],df['Thai']]
plt.boxplot(data,vert=True,widths=0.3,labels=["Math", "Eng", "Thai"],patch_artist=True)
plt.show()
```
