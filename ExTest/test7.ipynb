```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips =sns.load_dataset("tips")
```

```python
tips.head()
```

```python
plt.figure(figsize=[4,2])
sns.histplot(tips['tip'])
```

```python
plt.figure(figsize=[4,2])
sns.histplot(tips['total_bill'],bins = 20,kde = True)
```

```python
plt.figure(figsize=[4,2])
sns.countplot(data = tips, x = 'time')
```

```python
plt.figure(figsize=[4,2])
sns.countplot(data = tips, x = 'day', palette='magma')
```

```python
plt.figure(figsize=[4,2])
sns.histplot(tips['sex'])
```

```python
sns.boxplot(data=tips, x = 'day', y = 'tip')
```

```python
sns.boxplot(data=tips, y = 'day', x = 'tip')
```

```python
sns.boxplot(data=tips)
```

```python
sns.boxplot(data=tips, x = 'day', y = 'tip', hue='sex')
```

```python
sns.violinplot(data=tips, x = 'day', y = 'tip')
```

```python
sns.swarmplot(data=tips, x = 'day', y = 'tip', hue='sex')
```

```python
sns.scatterplot(data=tips, x = 'total_bill', y = 'tip', hue='sex')
```

```python
sns.lineplot(data=tips, x = 'total_bill', y = 'tip')
```

```python
sns.jointplot(data=tips, x = 'total_bill', y = 'tip', hue='sex')
```

```python
sns.jointplot(data=tips, x = 'total_bill', y = 'tip', kind='hex')
```

```python
sns.pairplot(data=tips)
```

```python
sns.pairplot(data=tips, hue='sex')
```

```python
g = sns.JointGrid(data=tips, x = 'total_bill', y = 'tip')
g.plot(sns.scatterplot, sns.boxplot)
```

```python
g = sns.PairGridGrid(data=tips, hue='sex')
g.map_diag(sns.scatterplot)
g.map_upper(sns.lineplot)
g.map_lower(sns.histplot)
```
