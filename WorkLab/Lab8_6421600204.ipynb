```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('World University Rankings 2023.csv',encoding = 'iso-8859-1')
```

```python
df.info()
```

```python
df.head()
```

```python
df2 = df[0:199]
```

```python
sns.scatterplot(data = df2 , x = 'Research Score', y = 'No of student per staff')
```

```python
df3 = df2[["Research Score","No of student per staff"]].dropna()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4 , random_state=0)
model.fit(df3)
```

```python
model.cluster_centers_
```

```python
plt.figure(figsize=[4,3])
sns.scatterplot(data = df3 , x = 'Research Score', y = 'No of student per staff'
                ,hue=model.labels_,palette='Set2')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1]
            ,color = 'k',marker='o')
```

```python
model.predict([[12,93],[20,64]])
```
